```python
parameter = {
    "dataset": "USStockPrice",
    "data_id": "AAPL",
    "start_date": "2020-01-01",
    "end_date": "2023-09-30",
    "token": "eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJkYXRlIjoiMjAyMy0xMC0wMyAxOTo1MDowMSIsInVzZXJfaWQiOiJrZXJyeTQxMDE1MTA0IiwiaXAiOiIxNDAuMTEzLjEzNi4yMTMifQ.YIrd6zarBavbYiZvRZjav6VetEAbA7-Y6_4dsXZE10U", 
}
data = requests.get(url, params=parameter)
data = data.json()
data = pd.DataFrame(data['data'])
data["date"] = pd.DatetimeIndex(data["date"])
data = data.set_index("date")
us_stock = data[["Open","High","Low","Close","Volume"]]
```

```python
parameter = {
     "dataset": "TaiwanExchangeRate",
     "data_id": "USD",
     "start_date": "2020-01-01",
     "end_date": "2023-09-30",
}
data = requests.get(url, params=parameter)
data = data.json()
data = pd.DataFrame(data['data'])
data["date"] = pd.DatetimeIndex(data["date"])
data = data.set_index("date")
ExchangeRate = data[['cash_sell']]
```

```python
api = DataLoader()
df = api.taiwan_stock_daily(
    stock_id='2330',
    start_date='2020-01-01',
    end_date='2023-09-30')
df["date"] = pd.DatetimeIndex(df["date"])
df = df.set_index("date")
tw_stock = df[["open","max","min","close","Trading_Volume"]]
tw_stock.rename(columns={'Trading_Volume':'Volume','open':'Open',"max":'High','min':'Low','close':'Close'}, inplace = True) 
```

```python
kwargs = dict(type='line', volume=True, figratio=(20,20), figscale=1.5, title="AAPL", style = 'binance')
mpf.plot(us_stock, **kwargs,savefig='AAPL.png')
```

```python
kwargs = dict(type='line', volume=True, figratio=(20,15), figscale=1.5, title="2330", style = 'binance')
mpf.plot(tw_stock, **kwargs,savefig='2330.png')
```
